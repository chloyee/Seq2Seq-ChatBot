```python
import matplotlib.pyplot as plt
import numpy as np

def plot_speech(audio, srate=16000):
    fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(20,10))
    time = np.arange(audio.shape[0])/srate
    ax1.plot(time, audio)
    ax1.set_xlim([0, time[-1]])
    ax2.specgram(audio, NFFT=512, Fs=srate, noverlap=128)
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
```

```python
import speech_recognition as sr
import glob
import os

def speech_to_text(sphfile, ambient_period=0):
    r = sr.Recognizer()
    audiofile = sr.AudioFile(sphfile)
    with audiofile as source:
        r.adjust_for_ambient_noise(source, duration=ambient_period)
        speech = r.record(source)
        return(r.recognize_google(speech))
        
```

```python
def str_to_bool(str):
    return True if str.lower() == 'true' else False
```

```python
"""
定义心跳检测函数
"""
def heartbeat():
    print (time.strftime('%Y-%m-%d %H:%M:%S - heartbeat', time.localtime(time.time())))
    timer = threading.Timer(60, heartbeat)
    timer.start()
# timer = threading.Timer(60, heartbeat)
# timer.start()
```

```python
# app = Flask(__name__,static_url_path="/static")
#
# @app.route('/message', methods=['POST'])

#"""定义应答函数，用于获取输入信息并返回相应的答案"""
def reply(text):
#从请求中获取参数信息
    # req_msg = request.form['msg']
    req_msg = text
#将语句使用结巴分词进行分词
    req_msg=" ".join(jieba.cut(req_msg))
    #调用decode_line对生成回答信息
    res_msg = execute.predict(req_msg)
    #将unk值的词用微笑符号袋贴
    res_msg = res_msg.replace('_UNK', '^_^')
    res_msg=res_msg.strip()
    
    # 如果接受到的内容为空，则给出相应的回复
    if res_msg == ' ':
        res_msg = '请与我聊聊天吧'

    return "人工智障：" + res_msg
    # return jsonify( { 'text': res_msg } )
```
